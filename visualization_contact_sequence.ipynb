```python
import pandas as pd
import numpy as np
import holoviews as hv
from holoviews import opts, dim
import math
from sklearn.preprocessing import StandardScaler
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from bokeh.io import export_png
hv.extension('bokeh')
print("module loaded")
```

```python
# directed user interaction flow

df = pd.read_csv("./transaction/19w/19w_profiled/clean_user_all.csv")
sequence_list = df["clean_app_sequence"].values
sequence_store_dict = {}
empty_sequence_count = 0
for sequence in sequence_list:
    sequence = eval(sequence)[1:]
    if len(sequence) == 0 or len(sequence) == 1:
        empty_sequence_count += 1
        continue
    i = 0
    for i in range(0, len(sequence)-1):
        slice_piece = tuple(sequence[i:i+2])
        if slice_piece in sequence_store_dict:
            sequence_store_dict[slice_piece] += 1
        else:
            sequence_store_dict[slice_piece] = 1

print(empty_sequence_count)

plot_list = []
for key, value in sequence_store_dict.items():
    temp_list = list(key)
    temp_list.append(value)
    plot_list.append(temp_list)

print(len(plot_list))
slice_num = 3
part_plot_list = sorted(plot_list,key=(lambda x:x[2]), reverse=True)[0:slice_num]

sankey = hv.Sankey(part_plot_list)
sankey.opts(width=1000, height=400, title=f"top {slice_num} directed user interaction flow")
```

```python
# undirected user interaction flow

df = pd.read_csv("./transaction/19w/19w_profiled/clean_user_all.csv")
sequence_list = df["clean_app_sequence"].values
sequence_store_dict = {}
empty_sequence_count = 0
for sequence in sequence_list:
    sequence = eval(sequence)
    if len(sequence) == 0 or len(sequence) == 1:
        empty_sequence_count += 1
        continue
    i = 0
    for i in range(0, len(sequence)-1):
        slice_piece = tuple(sequence[i:i+2])
        slice_piece_r = tuple(reversed(sequence[i:i+2]))

        if slice_piece in sequence_store_dict or slice_piece_r in sequence_store_dict:
            try:
                sequence_store_dict[slice_piece] += 1
            except:
                sequence_store_dict[slice_piece_r] += 1
        else:
            sequence_store_dict[slice_piece] = 1

print(empty_sequence_count)

plot_list = []
for key, value in sequence_store_dict.items():
    temp_list = list(key)
    temp_list.append(value)
    plot_list.append(temp_list)

print(len(plot_list))
slice_num = 40
part_plot_list = sorted(plot_list,key=(lambda x:x[2]), reverse=True)[0:slice_num]

sankey = hv.Sankey(part_plot_list)
sankey.opts(width=1000, height=400, title=f"top {slice_num} undirected user interaction flow")
```

```python
file_path = "./transaction/19w/19w_profiled/clean_user_tocate_expand_all.csv"
attribute = 'all'

translation_df = pd.read_csv('./contract_db/database.csv')

def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    try:
        category = translation_df.loc[translation_df['title']==word, 'category'].values[0]
    except:
        return "hsl(1, 100%%, %d%%)" % 50
    if category == 'games':
        return "hsl(185, 100%%, %d%%)" % 50
    elif category == 'marketplaces':
        return "hsl(37, 100%%, %d%%)" % 50
    elif category == 'gambling':
        return "hsl(74, 100%%, %d%%)" % 50
    elif category == 'collectibles':
        return "hsl(111, 100%%, %d%%)" % 50
    elif category == 'social':
        return "hsl(148, 100%%, %d%%)" % 50
    elif category == 'other':
        return "hsl(1, 100%%, %d%%)" % 50
    elif category == 'defi':
        return "hsl(222, 100%%, %d%%)" % 50
    elif category == 'high-risk':
        return "hsl(259, 100%%, %d%%)" % 50
    elif category == 'exchanges':
        return "hsl(296, 100%%, %d%%)" % 50

df = pd.read_csv(file_path)
df = df[['clean_to_title']]
word_dict = {}
for index, row in df.iterrows():
    try:
        app_dict = eval(row["clean_to_title"])
    except:
        continue
    for key, value in app_dict.items():
        if key in word_dict.keys():
            word_dict[key] += value
        else:
            word_dict[key] = value
del word_dict['unknown_addr']
del word_dict['self']

word_df = pd.DataFrame(word_dict.items(), columns=['Title', 'Value'])

wordcloud = WordCloud(background_color="white",width=3000, height=1000).generate_from_frequencies(word_dict)
plt.imshow(wordcloud.recolor(color_func=grey_color_func, random_state=3))
wordcloud.to_file(f"./wordcloud_{attribute}.png")
plt.axis("off")
plt.show()
```
